```python
from ultralytics import YOLO

model = YOLO("/content/best (1).pt")
```

```python
import gradio as gr
import tempfile
import os
def detect_fire(input_file):
    if input_file is None:
        return "No file uploaded", None

    # Create temporary directory
    temp_dir = tempfile.mkdtemp()
    input_path = input_file

    # Run prediction
    results = model.predict(
        source=input_path,
        save=True,
        conf=0.4,
        project=temp_dir,
        name="output"
    )

    # Find output file
    output_folder = os.path.join(temp_dir, "output")
    output_files = os.listdir(output_folder)

    # Return first output file (image or video)
    for f in output_files:
        if f.endswith((".jpg", ".png", ".mp4", ".avi", ".mov")):
            return "Detection Completed", os.path.join(output_folder, f)

    return "Prediction failed", None


# Gradio UI
interface = gr.Interface(
    fn=detect_fire,
    inputs=gr.File(label="Upload Image or Video"),
    outputs=[
        gr.Textbox(label="Status"),
        gr.Video(label="Output Video")  # will auto-work for images too
    ],
    title="üî• Fire Detection YOLOv8",
    description="Upload an image or video to detect fire using a fine-tuned YOLOv8 model.",
)

interface.launch()
```
